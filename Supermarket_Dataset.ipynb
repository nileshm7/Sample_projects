```python
#find sale from monday to sunday for ala data
df['Date'].dt.dayofweek.value_counts() #dt.dayofweek->will consider 0:'monday' .....6:'sunday'
```

```python
#will map for 0 to 6 = mon to sundy
#note 1st check how many dates r there from data in this example 7 dates so  map forall 7dates
#ad new call
df['Day_of_week']=df['Date'].dt.dayofweek.map({0:'Monday',1:'Tuseday',2:'Wednesday',3:'Thrusday',4:'Friday',5:'Saturday',6:'Sunday'})
df['Day_of_week'].value_counts()
```

```python
df_weekwise=df['Day_of_week'].value_counts()
plt.pie(df_weekwise,labels=df_weekwise.index,autopct='%1.2f%%')
plt.show()


```

```python
#find sale for months from jan tp dec  of all data
#note 1st check how many months r there from data in this example 3 month so only map for 3 month
df['Date'].dt.month.value_counts() #dt.dayofweek->will consider 1=jan 2= feb 3=mar map for 1to3=jan to mar
```

```python
#will map for 1to3 = jan to mar
#ad new call
df['Mothswise_data']=df['Date'].dt.dayofweek.map({1:'Jan',2:'Feb',3:'Mar'})
df['Mothswise_data'].value_counts()
```

```python
df_mothwise=df['Mothswise_data'].value_counts()
plt.pie(df_mothwise,labels=df_mothwise.index,autopct='%1.2f%%')
plt.show()

```
